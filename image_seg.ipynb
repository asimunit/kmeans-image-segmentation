```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
im = cv2.imread("strange.jpg")
im =cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(im)
```

```python
X=im.reshape(-1,3)
print(X)
```

```python
model=KMeans(5)
```

```python
model.fit(X)
```

```python
model.labels_
```

```python
for rgb in model.cluster_centers_:
    im_t = np.zeros([100, 100, 3], dtype=int)
    im_t[:, :] = rgb
    plt.figure()
    plt.imshow(im_t)
    plt.show()
```

```python
potent_colors =  model.cluster_centers_
```

```python
res = np.zeros([X.shape[0], X.shape[1]], dtype=int)
```

```python

for i, lab in enumerate(model.labels_):
    res[i] = potent_colors[lab]
```

```python

res = res.reshape(im.shape[0], im.shape[1], im.shape[2])
```

```python

plt.imshow(res)
```
